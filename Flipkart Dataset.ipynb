```python
#calculate average rating
avg_rating=df['Ratings'].mean()
print(f'Average rating:{avg_rating}')
```

```python
#distribution of ratings 
plt.figure(figsize=(8,6))
sns.histplot(df['Ratings'],kde=True,bins=10)
plt.title('Distribution of rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# bar plot of average discount by brand
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Discount',data=df,estimator=np.mean)
plt.title('Average disount by brand')
plt.xlabel('Brand')
plt.ylabel('Average discount (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
#count plot of number of products by brand
plt.figure(figsize=(10,6))
sns.countplot(x='Brand', data=df)
plt.title('Number of products by brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# average rating of top 5 brands by count
top_brands=df['Brand'].value_counts().head(5).index
plt.figure(figsize=(10,6))
sns.barplot(x='Brand', y='Ratings', data=df[df['Brand'].isin(top_brands)], estimator=np.mean)
plt.title('Average rating of top 5 brands')
plt.xlabel('Brands')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#msp column to numeric
df['MSP']=pd.to_numeric(df['MSP'], errors='coerce')

df=df.dropna(subset=['MSP'])
#define
bins=[0,10000,20000,30000,50000,100000]
labels=['Low-end(<10K)','Budget(10K-20K)','Mid-range(20K-30K)','Primium(30K-50K)','Flagship(>50K)']

#create new column for price range
df['Price_Range']=pd.cut(df['MSP'],bins=bins,labels=labels)
# cost of mobile in each price range
price_range_count=df['Price_Range'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=price_range_count.index, y=price_range_count.values)
plt.title('Mobile count in different price segment in india')
plt.xlabel('Price range')
plt.ylabel('Number of mobiles')
#adjust layout to avoide overlap
plt.tight_layout()
plt.show()
```

```python
#group by brand and get max discount for each
brand_highest_discount=df.groupby('Brand')['Discount'].max().reset_index()
top_brands_by_discount=brand_highest_discount.sort_values(by='Discount',ascending=False).head(5)
plt.figure(figsize=(10,6))
sns.barplot(x='Brand', y='Discount', data=top_brands_by_discount,palette='coolwarm', hue='Brand',legend=False)
plt.title(' brand and get max discount for each')
plt.xlabel('Brands')
plt.ylabel('discountg')
plt.tight_layout()
plt.show()
```

```python

df.groupby('Brand')['Ratings'].mean().nlargest(5).plot(kind='bar')
plt.title(' Top 5 brands by average rating')
plt.ylabel('average rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#top 5 brands by total no of review
a=df.groupby('Brand')['No_of_reviews'].sum().sort_values(ascending=False).head(5)
b=a.reset_index()
sns.barplot(x='Brand', y='No_of_reviews', data=b,palette='viridis', hue='Brand',legend=False)
plt.title('Top 5 brands by total no of review')
plt.xlabel('Brands')
plt.ylabel('Total no of review')
plt.tight_layout()
plt.show()
```
